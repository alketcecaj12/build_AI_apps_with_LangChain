```python
## import langchain and openAI libs 
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

import os
from dotenv import load_dotenv
load_dotenv()

# Define the LLM
llm = ChatOpenAI(
    model="gpt-4o-mini",
    api_key=os.getenv("OPENAI_API_KEY")  # ✅ reads from .env file
)
```

```python
# Create a prompt template from the template string
template = "You are an artificial intelligence assistant, answer the question. {question}"
prompt = PromptTemplate.from_template(
    template=template
)

chain = prompt | llm | StrOutputParser()  # ✅ extracts just the text

# Invoke the chain on the question

response = chain.invoke({"question": 
                        "How does LangChain make LLM application development easier?"})
print(response)
```
