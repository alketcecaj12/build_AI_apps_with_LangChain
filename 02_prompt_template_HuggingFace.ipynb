#### Using HuggingFace and Llama3.2 non reccomended

#### HuggingFace Hub 
You can load llama3.2 via HuggingFace, but:

- You need to request access from Meta on HuggingFace (approval required)
- The full model is ~6GB+ to download
- Requires a HuggingFace account + token

## Final decision : too long and cumbersome process
