```python
# ✅ Fix 1: Add the missing import
from langchain_openai import ChatOpenAI

# ✅ Fix 2: Use a real API key — ideally from .env, not hardcoded
import os
from dotenv import load_dotenv
load_dotenv()

# Define the LLM
llm = ChatOpenAI(
    model="gpt-4o-mini",
    api_key=os.getenv("OPENAI_API_KEY")  # ✅ reads from .env file
)

# Predict the words following the text in question
prompt = 'Three reasons for using LangChain for LLM application development.'
response = llm.invoke(prompt)

print(response.content)
```
